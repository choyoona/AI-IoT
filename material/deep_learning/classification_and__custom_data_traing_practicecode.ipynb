```python
# 남이 잘 만들어 놓은 특질 추정 부분은 그대로 사용하고, 내가 원하는 부분만을 학습시키는 것을 전이학습 이라고 한다. 

model = keras.Sequential)()
model.add(Input(224,224,3)) 
model.add(Conv2D(64, (3,3), padding='same'))
model.add(Conv2D(64, (3,3), padding='same'))   # padding을 same으로 하면 바뀌지 않는다
model.add(MaxPooling2D((2,2))   
model.add(Conv2D(128, (3,3), padding='same')
model.add(Conv2D(128, (3,3), padding='same')
model.add(MaxPooling2D((2,2))  
model.add(Conv2D(512, (3,3), padding='same')
model.add(Conv2D(512, (3,3), padding='same')
model.add(Conv2D(512, (3,3), padding='same')
model.add(MaxPooling2D((2,2))
model.add(Conv2D(512, (3,3), padding='same')
model.add(Conv2D(512, (3,3), padding='same')
model.add(Conv2D(512, (3,3), padding='same')
model.add(MaxPooling2D((2,2))
model.add(Flatten())
model.add(Dense(4069, activation="relu"))
model.add(Dense(4069, activation="relu"))
model.add(Dense(1000, activation="softmax"))    # 아래와 같은 사진을 코딩으로 재현한 것 
```
